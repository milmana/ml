```python
import numpy as np
import matplotlib.pyplot as plt
y = lambda x: .5*x + 3
N=10
x_ = np.linspace(0,1, N)
y_ = np.array([y(x) for x in x_])
plt.scatter(x_, y_, marker='x', c='r')
```

```python
rng = np.random.default_rng()
x_train = x_ + rng.normal(scale=0.1,size=x_.size)
y_train = y_ + rng.normal(scale=0.1,size=y_.size)
plt.scatter(x_train, y_train, marker='o', c='b')
```

```python
class Pred:
    def __init__(self, w, b):
        self.w_ = w
        self.b_ = b
    def __call__(self, x, y):
        w = self.w_
        b = self.b_
        p = np.array([w*x_i + b for x_i in x])
        e = np.array([(p_i-y_i)*(p_i-y_i) for p_i, y_i in zip(p,y)])
        e = e/len(e)
        return p, e

pred = Pred(.5,3)
y_pred, mse = pred(x_train, y_train)
fig, ax = plt.subplots()
ax.plot(x_train, y_pred, c='r')
ax.scatter(x_train, y_train, c='b')
plt.show()
```
